# Evaluation in a Clinical Setting – A RevCycle Use Case 🚀⚕️

_Evaluating clinical cases for prior authorization is a critical part of the revenue cycle (RevCycle) in healthcare. This process ensures that patients receive the appropriate treatments while maintaining cost-effective practices and compliance with policies._

## Overview 📋

Prior authorization decisions require the integration of multiple clinical contexts. In our RevCycle use case, decision-making is informed by:

- **Patient Information** 👤  
  Data such as patient demographics, medical history, and contact details provide the foundation for understanding the patient’s background.

- **Physician Information** 👩‍⚕️  
  Information about the prescribing physician, including their specialty, credentials, and contact information, is essential for verifying clinical decisions and ensuring adherence to treatment protocols.

- **Clinical Information** 🩺  
  Detailed clinical data—including diagnoses, lab results, imaging, and treatment histories—outline the patient’s current clinical status and the urgency of care. This context ensures that decisions are medically justified and tailored to the patient’s condition.

- **Policy Supporting** 📜  
  Policy text and guidelines play an integral role in evaluating if the request meets the required criteria for prior authorization. These policies ensure that clinical decisions are aligned with external standards and reimbursement rules.

## How It Works 🔄

1. **Data Collection and Contextualization**  
   The process begins by collecting data from various sources, including patient and physician records, lab reports, and imaging results. This data is then structured into defined contexts:
   - **Patient Context:** Captures key patient demographics and history.  
   - **Physician Context:** Documents the credentials and specialty of the healthcare provider.  
   - **Clinical Context:** Aggregates the clinical findings, treatment history, lab results, and imaging.  
   - **Policy Context:** Contains the guidelines and policy text that support prior authorization decisions.

2. **Integration and Evaluation**  
   An evaluation pipeline uses these contexts to generate clinical decisions. Advanced language models and evaluators assess:
   - **Factual Correctness**  
   - **Semantic Similarity**  
   - **Fuzzy Matching of Clinical Criteria**  

   This ensures that all relevant criteria from the policy are met, and that the clinical reasoning is robust. The integration of these diverse contexts allows the evaluator to comprehensively validate the medical necessity and appropriateness of the request.

3. **Continuous Improvement and Collaboration**  
   By using a declarative framework (e.g., YAML-based test case specifications), both technical and clinical domain experts can define and update test cases. This collaborative process promotes rapid iteration and consistent decision-making across revisions, making it easier to adjust the criteria as policies and clinical guidelines evolve.

## Benefits for Revenue Cycle Management 💡

- **Efficiency:** Streamlined data integration minimizes manual review and accelerates the authorization process.  
- **Accuracy:** Combining multiple clinical contexts leads to more precise and defensible prior authorization decisions.  
- **Compliance:** Automated matching against policy criteria ensures adherence to reimbursement rules.  
- **Collaboration:** A declarative test case framework empowers clinicians and policy experts to directly contribute to the evaluation process without requiring deep programming knowledge.

## 4. Evaluators: Validating Decisions with Multiple Metrics

Our robust evaluation pipeline employs several evaluators to analyze both the **deterministic decision** and the **non-deterministic rationale** generated by the module. These evaluators include:

- **FuzzyEvaluator**: Checks for string-level similarity.  
- **SemanticSimilarityEvaluator**: Assesses semantic alignment between rationale texts.  
- **GPT Similarity Evaluator**: Leverages advanced NLP backed by LLMs to compare detailed reasoning.  
- **FactualCorrectnessEvaluator**: Verifies the factual accuracy of responses using LLMs.  
- **F1ScoreEvaluator (AI Foundry)**: Performs an F1-based scoring of discrete classification or token-level matches.

---

### FactualCorrectnessEvaluator

This metric compares and evaluates the factual accuracy of the generated response against the reference. It leverages a RAG-as-based approach to break down both the response and reference into individual claims and then uses natural language inference (NLI) to determine factual overlap between them.

- **True Positive (TP):** claims in both response and reference  
- **False Positive (FP):** claims in response but not in reference  
- **False Negative (FN):** claims in reference but not in response  

Overall factual correctness ranges from 0 to 1, with higher values indicating better alignment.

---

### Deterministic vs. LLM-as-a-Judge Approaches

|                             | **Deterministic Evaluators**              | **LLM-as-a-Judge Evaluators**            |
|-----------------------------|-------------------------------------------|------------------------------------------|
| **Pros**                    | - Consistent, reproducible<br>- Easy to debug and trace<br>- Lower cost per eval | - Understand subtle context<br>- Handle paraphrase variation<br>- Apply domain reasoning |
| **Cons**                    | - May miss nuanced language<br>- Rigid thresholds<br>- Requires tuning | - More expensive/slower<br>- Less transparent logic<br>- Potential variability/bias |

This multi-pronged strategy ensures subtle errors are detected while balancing speed, cost, and depth of understanding.

````python
# Cell 8: Fuzzy string-level similarity
from src.evals.custom.fuzzy_evaluator import FuzzyEvaluator

fuzzy_eval = FuzzyEvaluator()

# perfect match
print(fuzzy_eval(response=result['Determination'], ground_truth="Approved"))
# mismatched
print(fuzzy_eval(response=result['Determination'], ground_truth="Rejected"))
# case-insensitive
print(fuzzy_eval(response=result['Determination'], ground_truth="approved"))

# more complex: did all policy criteria fully meet?
count_not_fully = len([
    x for x in result['DetailedAnalysis']['PolicyCriteriaAssessment']
    if x['Assessment'] != 'Fully Met'
])
print(fuzzy_eval(response=str(count_not_fully), ground_truth="0"))
````

## 7. Framework for Declarative Test-Case Specification
Instead of hand-coding every scenario, we can let **domain experts** write **YAML** (or JSON) test-cases that drive the same evaluators:

**Benefits of a Declarative Approach**
- **Accessible to Non-Developers**  
  Clinical experts or policy analysts can add/modify tests without touching Python.  
- **Rapid Iteration & Reproducibility**  
  Versioned schemas + config files freeze inputs & outputs.  
- **Consistent Metrics**  
  All evaluators read the same structured data.  
- **Collaboration Across Disciplines**  
  Everyone speaks the same “contract” (YAML), from docs to code.
