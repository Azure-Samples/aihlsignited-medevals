# Evaluation in a Clinical Setting - A RevCycle Use Case 🚀🏥

Evaluating clinical cases for prior authorization is a critical part of the revenue cycle (RevCycle) in healthcare. This process ensures that patients receive the appropriate treatments while maintaining cost-effective practices and compliance with policies.

## Overview 📋
Prior authorization decisions require the integration of multiple clinical contexts. In our RevCycle use case, decision-making is informed by:

- **Patient Information** 🧑‍⚕️  
  Data such as patient demographics, medical history, and contact details provide the foundation for understanding the patient's background.

- **Physician Information** 👨‍⚕️  
  Information about the prescribing physician, including their specialty, credentials, and contact information, is essential for verifying clinical decisions and ensuring adherence to treatment protocols.

- **Clinical Information** 🩺  
  Detailed clinical data, including diagnoses, lab results, imaging, and treatment histories, outline the patient’s current clinical status and the urgency of care. This context ensures that decisions are medically justified and tailored to the patient's condition.

- **Policy Supporting** 📜  
  Policy text and guidelines play an integral role in evaluating if the request meets the required criteria for prior authorization. These policies ensure that clinical decisions are aligned with external standards and reimbursement guidelines.

## How It Works 🔄

1. **Data Collection and Contextualization**  
   The process begins by collecting data from various sources, including patient and physician records, lab reports, and imaging results. This data is then structured into defined contexts:
   - **Patient Context:** Captures key patient demographics and history.
   - **Physician Context:** Documents the credentials and specialty of the healthcare provider.
   - **Clinical Context:** Aggregates the clinical findings, treatment history, lab results, and imaging.
   - **Policy Context:** Contains the guidelines and policy text that support prior authorization decisions.

2. **Integration and Evaluation**  
   An evaluation pipeline uses these contexts to generate clinical decisions. Advanced language models and evaluators assess:
   - **Factual Correctness**
   - **Semantic Similarity**
   - **Fuzzy Matching of Clinical Criteria**

   This ensures that all relevant criteria from the policy are met, and that the clinical reasoning is robust. The integration of these diverse contexts allows the evaluator to comprehensively validate the medical necessity and appropriateness of the request.

3. **Continuous Improvement and Collaboration**  
   By using a declarative framework (e.g., YAML-based test case specifications), both technical and clinical domain experts can define and update test cases. This collaborative process promotes rapid iteration and consistent decision-making across revisions, making it easier to adjust the criteria as policies and clinical guidelines evolve.

## Benefits for Revenue Cycle Management 💡
- **Efficiency:** Streamlined data integration minimizes manual review and accelerates the authorization process.
- **Accuracy:** Combining multiple clinical contexts leads to more precise and defensible prior authorization decisions.
- **Compliance:** Automated matching against policy criteria ensures compliance with reimbursement rules.
- **Collaboration:** A declarative test case framework empowers clinicians and policy experts to directly contribute to the evaluation process without requiring deep programming knowledge.

## 4. Evaluators: Validating Decisions with Multiple Metrics

Our robust evaluation pipeline employs several evaluators to analyze both the **deterministic decision** and the **non-deterministic rationale** generated by the module. These evaluators include:
 
- **FuzzyEvaluator:** Checks for string-level similarity.  
- **SemanticSimilarityEvaluator:** Assesses semantic alignment between rationale texts.  
- **GPT Similarity Evaluator:** Leverages advanced NLP backed by LLMs to compare detailed reasoning.  
- **FactualCorrectnessEvaluator:** Verifies the factual accuracy of responses using LLMs
- **F1ScoreEvaluator (AI Foundry):** Performs an F1-based scoring of discrete classification or token-level matches.

**FactualCorrectnessEvaluator**  
This metric compares and evaluates the factual accuracy of the generated response against the reference. It leverages a RAGAS-based approach to break down both the response and reference into individual claims and then uses natural language inference (NLI) to determine factual overlap between them. The overall factual correctness ranges from 0 to 1, with higher values indicating better alignment. Specifically, the evaluator computes **precision**, **recall**, and **F1** scores for these claims, which can be adapted using a mode parameter to emphasize certain types of matches.

- **True Positive (TP):** The number of factual claims the response and reference both share.  
- **False Positive (FP):** The number of factual claims that exist in the response but are not in the reference.  
- **False Negative (FN):** The number of factual claims missing in the response but present in the reference.

**Deterministic Evaluators vs. LLM-as-a-Judge Approaches**

- **Deterministic Evaluators** (e.g., string-based similarity, semantic embeddings, factual checks):
  - **Pros**: 
    - They provide *consistent and reproducible* results since the logic is rule-based or statistically based on embeddings. 
    - They are *easier to debug* and trace because each step of the scoring or comparison can be followed explicitly (e.g., fuzzy string matching, exact or partial overlaps). 
    - They do not introduce additional complex language model dependencies, so the cost per evaluation can be lower.
  - **Cons**: 
    - They may *overlook nuanced language differences* or more sophisticated forms of reasoning. 
    - They can be *too rigid* for open-ended or highly varied responses because they rely heavily on predetermined metrics or thresholds.
    - They often require *careful tuning of thresholds* or embeddings to avoid false positives or negatives.

- **LLM-as-a-Judge Evaluators** (e.g., GPT Similarity Evaluator):
  - **Pros**: 
    - They can *understand subtle context and reasoning* in an open-ended, more human-like manner. 
    - They handle *complex language variations* and paraphrases effectively, capturing meaning beyond simple string or embedding overlaps. 
    - They can apply *domain reasoning* if the underlying model has been trained on relevant data, making it more adaptive to complex or nuanced tasks.
  - **Cons**: 
    - They can be *more expensive to run* (in terms of compute resources or API calls) and *potentially slower* for large-scale batch evaluations. 
    - Their judgments can be *less transparent* or interpretable, as the evaluation logic is buried within the model’s parameters. 
    - They can introduce *bias* or *variability* in judgments if the LLM itself exhibits inconsistencies.

This multi-pronged strategy ensures our system detects subtle errors and maintains high-quality outputs. Using a combination of deterministic metrics and advanced LLM-based comparisons offers the best coverage across scenarios—deterministic evaluators deliver consistency, while LLM-as-a-Judge approaches provide adaptability and deeper context understanding. 📊

Are you seeing the pattern now? While our ground truth scores very high with cosine difference, imagine being in operations understanding the difference in scoring when this is deployed into production. Our metrics are only as strong as the _explainability_ behind them, and so if we select this as a metric and our model doesn't exhaustively match the response, are we okay with that as a justification for a low score?

AI Foundry is our toolbox, and from the toolbox we can pick from the tools that come with it or use the toolbox to build our own with what is out there. This introduces us to a new evaluator which is available in the open source community from a popular framework, RAGAS.

RAGAS’s **Factual Correctness** metric turns your LLM into its own referee: it splits each answer into discrete factual claims, asks a judge-LLM to check whether every claim is *entailed* by the retrieval context with an NLI-style prompt, and then rolls the per-claim verdicts into a 0-to-1 score that shows how many statements the model got right. This claim-level test catches hallucinations that word-overlap metrics miss and slots neatly into CI/CD so every pull-request can be gated on “no new factual errors.”

- **Drop-in usage** – `from ragas.metrics import FactualCorrectness` → `fc = FactualCorrectness()` → `fc.score(dataset)`; runs offline with your chosen judge model.  
- **Tunable judge** – swap the default prompt, pick a domain-tuned LLM, or fine-tune on expert annotations to tighten or relax what counts as “supported.”  
- **Pipeline-ready signals** – returns both the aggregate score and the list of unsupported claims so dashboards or alerting rules can act on concrete failure cases.

Let's give it a shot:

````python
fact_corr = FactualCorrectnessEvaluator(model_config=model_config)
print(fact_corr(
    response=result['Rationale'],
    ground_truth=(
        "Based on the provided information, Sarah Sample meets all the policy criteria "
        "for initial therapy approval of adalimumab for Crohn's Disease. She is over 6 years old, "
        "has tried corticosteroid therapy without improvement, and the medication is prescribed by a pediatric gastroenterologist."
    )
))
````

````python
print(fact_corr(response="Sarah is rejected.", ground_truth=(
        "Based on the provided information, Sarah Sample meets all the policy criteria "
        "for initial therapy approval of adalimumab for Crohn's Disease. She is over 6 years old, "
        "has tried corticosteroid therapy without improvement, and the medication is prescribed by a pediatric gastroenterologist."
    )))
````

````python
print(fact_corr(response="Based on the requirements, Sarah meets them is approved for the policy.", ground_truth=(
        "Based on the provided information, Sarah Sample meets all the policy criteria "
        "for initial therapy approval of adalimumab for Crohn's Disease. She is over 6 years old, "
        "has tried corticosteroid therapy without improvement, and the medication is prescribed by a pediatric gastroenterologist."
    )))
````

Note that with the negative scenario, the LLM scored a complete zero while for the minimal positive scenario it's an incremental value above 0. This aligns well with how we wish to operationalize the evaluation. The LLM should be able to identify the positive scenario with a higher score, while the negative scenario should be scored as zero or near-zero.

The typical workflow is:

1. **Prepare your data**: In this example, we write a JSONL file (`temp_dataset`) containing the scenarios we want to test.  
2. **Configure your evaluators**: We define each evaluator’s mapping—i.e., how it reads the fields from the dataset—and any additional parameters needed.

4. **Call the `evaluate` function**: We pass our evaluation data, evaluator configuration, and Foundry project settings to the `evaluate` function.  
5. **View and track results**: The metrics generated (e.g., fuzzy similarity scores, factual correctness, or GPT-based judgments) are stored in Foundry’s database or UI, allowing you to compare runs across time or different model versions.

## 8. Framework for Declarative Test-Case Specification

Instead of writing custom Python code for every evaluation scenario, our framework leverages a declarative format (such as YAML or JSON) to define test cases. Test cases are specified in configuration files that detail the context (such as patient, physician, clinical data, and policy text) as well as the evaluators to apply and the expected results or metrics. Please reference the lab notebook (part 3) to learn more behind why we do this, and why it is important for us to think about this to scale.

### Benefits of a Declarative Approach

- **Accessible to Non-Developers:**  
  By using a plain-text format like YAML, clinical experts, policy analysts, and other domain specialists can directly write and modify test cases. This democratizes the process, meaning that contributions are not limited solely to Python developers.

- **Rapid Iteration & Reproducibility:**  
  Declarative test cases enable easy versioning and modification across commits. New test scenarios can be added without needing to alter the underlying code, which leads to faster iterations and more reliable results. Since the specification and expected outcomes are clearly defined in the configuration, reproducibility is improved.

- **Consistent Evaluation Metrics:**  
  When the evaluators are plugged in to read from the same structured data, consistency in measuring metrics (such as fuzzy similarity, factual correctness, and semantic similarity) is enforced. Changes to evaluation criteria (for example, tweaking thresholds) can be made by simply updating a YAML file.

- **Collaboration Across Disciplines:**  
  Domain experts can define test case scenarios in a familiar format without requiring deep knowledge of the codebase. They can, for instance, document contextual information and expected outcomes in language that aligns with clinical guidelines or policy requirements. The technical team then builds a parser that converts these specifications into runtime test cases.
